Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Observing our Dataset

```python
df = pd.read_csv('human_development_index_vs_mean_male_height.csv')
df
```

Describing our Dataset (rounded to 3 decimal points, excluding the 'Year' column)

```python
hdi_height = df.copy() # make a copy so we can go back to the original dataset if needed

hdi_height_1 = hdi_height.drop(columns=['Year'])
hdi_height_1.describe().round(3)
```

Data Cleaning: Dropping Null Values

```python
hdi_height_cleaned = hdi_height.dropna()
hdi_height_cleaned
```

Describing our Cleaned Dataset (rounded to 3 decimal points, excluding the 'Year' column)

```python
hdi_height_new = hdi_height_cleaned.drop(columns=['Year'])
hdi_height_new.describe().round(3)
```

Finding the Correlation between Both Variables

```python
correlations = hdi_height_new.corr()
correlations['Human Development Index (UNDP)']
```

Data Visualization (with and without line of best fit)

```python
x = 'Human Development Index (UNDP)'
y = 'Mean male height (cm)'

hdi_height_new.plot.scatter(x, y, title='Scatterplot of HDI and Mean Height');

sns.lmplot(x, y, data=hdi_height_new)
plt.title("Scatterplot of HDI and Male Height");
```
