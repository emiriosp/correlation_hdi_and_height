```python
import pandas as pd
```

```python
df_age_gender = pd.read_csv('age_gender.csv')
df_edu = pd.read_csv('edu_level.csv')
df_migration = pd.read_csv('location_merged.csv')
df_rural = pd.read_csv('rural.csv')
df_smoking = pd.read_csv('smoking.csv')
df_smoking_behavior = pd.read_csv('smoking_merged.csv')

display(df_age_gender,df_edu,df_migration,df_rural,df_smoking,df_smoking_behavior)
```

```python
df_edu_new = df_edu.sort_values('edu_level', ascending=False).drop_duplicates('pidlink').sort_index()

df_edu_new
```

```python
df0 = pd.merge(df_age_gender,df_edu_new,how='outer',on='pidlink')
df1 = pd.merge(df_migration,df_smoking_behavior,how='outer',on='pidlink')
df2 = pd.merge(df0,df1,how='outer',on='pidlink')
df3 = pd.merge(df_rural,df_smoking,how='outer',on='hhid14_9')
df4 = pd.merge(df2,df3,how='outer',on='pidlink')

df5 = df4[['pidlink','age','male','edu_level','rural','smoking','rural_no_migration','urban_no_migration','urban_to_rural_migration',
'rural_to_urban_migration','rural_to_rural_migration','urban_to_urban_migration','never_smoking','always_smoking','started_smoking']].sort_index()

df5
```

```python
df5.to_csv(r'C:\data_python\all_merged.csv',index=False)
```
