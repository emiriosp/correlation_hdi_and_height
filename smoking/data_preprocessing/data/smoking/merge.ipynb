```python
import pandas as pd
```

```python
df0 = pd.read_csv('smoking_original.csv')
df1 = pd.read_csv('lagged_smoking_original.csv')

display(df0,df1)
```

```python
df2 = pd.merge(df0,df1,how='left',on='pidlink').dropna()

df2
```

```python
def smoking(row):
    if row['smoking']-row['lagged_smoking'] == 0 and row['smoking'] == 0:
        return 1
    return 0
df2['never_smoking'] = df2.apply(smoking, axis=1)

def smoking(row):
    if row['smoking']-row['lagged_smoking'] == 0 and row['smoking'] == 1:
        return 1
    return 0
df2['always_smoking'] = df2.apply(smoking, axis=1)

def smoking(row):
    if row['smoking']-row['lagged_smoking'] == 1:
        return 1
    return 0
df2['started_smoking'] = df2.apply(smoking, axis=1)

def smoking(row):
    if row['smoking']-row['lagged_smoking'] == -1:
        return 1
    return 0
df2['stopped_smoking'] = df2.apply(smoking, axis=1)
```

```python
df3 = df2[['pidlink','never_smoking','always_smoking','started_smoking','stopped_smoking']]

df3
```

```python
df3.to_csv(r'C:\data_python\smoking_merged.csv',index=False)
```
