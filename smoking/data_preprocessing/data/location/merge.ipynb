```python
import pandas as pd
```

```python
df0 = pd.read_csv('rural_original.csv')
df1 = pd.read_csv('lagged_rural_original.csv')
df2 = pd.read_csv('migration_original.csv')

display(df0,df1,df2)
```

```python
df3 = pd.merge(df0,df1,how='outer',left_on='hhid14_9',right_on='hhid07_9')
df4 = df3.drop(columns='hhid07_9')

df5 = pd.merge(df4,df2,how='outer',on='hhid14_9')
df6 = df5.drop(columns='hhid14_9').dropna()

df6
```

```python
def migration(row):
    if row['migrated_2007_2014'] == 0 and row['rural'] == 1:
        return 1
    return 0
df6['rural_no_migration'] = df6.apply(migration, axis=1)

def migration(row):
    if row['migrated_2007_2014'] == 0 and row['rural'] == 0:
        return 1
    return 0
df6['urban_no_migration'] = df6.apply(migration, axis=1)

def migration(row):
    if row['migrated_2007_2014'] == 1 and row['rural']-row['lagged_rural'] == 1:
        return 1
    return 0
df6['urban_to_rural_migration'] = df6.apply(migration, axis=1)

def migration(row):
    if row['migrated_2007_2014'] == 1 and row['rural']-row['lagged_rural'] == -1:
        return 1
    return 0
df6['rural_to_urban_migration'] = df6.apply(migration, axis=1)

def migration(row):
    if row['migrated_2007_2014'] == 1 and row['rural']-row['lagged_rural'] == 0 and row['rural'] == 1:
        return 1
    return 0
df6['rural_to_rural_migration'] = df6.apply(migration, axis=1)

def migration(row):
    if row['migrated_2007_2014'] == 1 and row['rural']-row['lagged_rural'] == 0 and row['rural'] == 0:
        return 1
    return 0
df6['urban_to_urban_migration'] = df6.apply(migration, axis=1)
```

```python
df7 = df6[['pidlink','rural_no_migration','urban_no_migration','urban_to_rural_migration',
'rural_to_urban_migration','rural_to_rural_migration','urban_to_urban_migration']]

df7
```
